import libraries, create bar graph for Top U.S. states with the highest percentage of poor mental health among adults 2018

```python

import numpy as np 
import matplotlib.pyplot as plt  
import seaborn as sns
import pandas as pd
rates = pd.read_csv('rates.csv')
sns.set(style="whitegrid")
f, ax = plt.subplots(figsize=(10, 8))
sns.barplot(x="Percent", y="State", linewidth=10, data=rates,color="#3498db").set_title('Top U.S. States with the Highest Percentage of Poor Mental Health Among Adults', fontname='Arial', fontsize=24)
ax.set(xlim=(38.5, 42))
sns.despine(left=True, bottom=True, top = True, right = True)
```

create bar plot for Reasons for not receiving mental health services among U.S. adults 2018

```python
treatment=pd.read_csv('treatment.csv')
sns.set(style="whitegrid")
f, ax = plt.subplots(figsize=(12, 8))
sns.barplot(x="Percent", y="Reason", linewidth=10, data=treatment,color="#3498db").set_title('Reasons for not receiving mental health services among U.S. adults 2018', fontname='Arial', fontsize=24)
ax.set(xlim=(0, 42))
sns.despine(left=True, bottom=True, top = True, right = True)
```
